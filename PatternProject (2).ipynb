## Setting Global values for aplha,intercept,and others


```python
alphas = clf.dual_coef_
support_vectors = clf.support_vectors_
bias = clf.intercept_
Ns = len(support_vectors)
print ('Aplhas',alphas)
print ('Support Vectors',support_vectors)
print ('Bias',bias)
print ('Ns', Ns)
```

```python
# Creating test set
import pickle
pickle_in = open("test_features (1).pkl","rb")

testSet = pickle.load(pickle_in)
testSet = np.array(testSet, dtype=float)
```

```python
# Helper function
def getX(ci):
    res = np.array(data.loc[data[data.columns[len(data.columns)-1]] == ci], dtype= float)
    # print ("Delta", res[0]-res[1])
    return res[0,:-1]
```

```python
# Helper function
def getX2(ci):
    res = np.array(data.loc[data[data.columns[len(data.columns)-1]] == ci], dtype= float)
    # print ("Delta", res[0]-res[1])
    return res[1,:-1]
```

```python
def subtract(a,b):
    a= np.array(a, dtype =float)
    b = np.array(b,dtype = float)
    return np.absolute(a - b)
```

```python
def computeKernal(a,sub):
    sub = sub.reshape(-1,1)
    return (a) @ (sub)
```

```python
'''
Terminologies:
1. probe = test image
2. ci = claimed identity
'''
def verification(probe,ci):
    # x_ = extract_proble_features(probe)
    x = getX(ci)
    sub = subtract(x,probe)
    similarity = 0
    for i in range(Ns):
        alpha = alphas[0][i]
        similarity = similarity + (alpha * computeKernal(support_vectors[i],sub))
    return similarity
     
```

# Calculation of threshold

```python
threshold = 0
for i in range(20):
    i=i+1
    thr = getX2(i)
    re = verification(thr,i)
    threshold = threshold+ re
threshold = (threshold/20) * 1.3 #1.1 is the bias factor 
print (threshold)
```

# Results of verification

```python
claimed ={"false positive":0,
         "false negative": 0,
         "true positive":0,
         "true negative":0}
def score(testSet,ci, threshold):
    input = testSet[:-1]
    output = testSet[-1]
    if(verification(input,ci) < threshold):
        if(ci == output):
            tp = claimed.get("true positive")
            claimed.update({"true positive":tp+1})
        else:
            fp = claimed.get("false positive")
            claimed.update({"false positive":fp+1})

        return output,"Claim Accepted"
    else:
        if(ci == output):
            fn = claimed.get("false negative")
            claimed.update({"false negative":fn+1})
        else:
            tn = claimed.get("true negative")
            claimed.update({"true negative":tn+1})
        
        return output,"Claim rejected"
```

```python
dict_results= dict()

for i in range(10,20):
    i =i+1
    ans =dict()
    results = np.apply_along_axis(score,1,testSet,i,threshold)
    for k in range(results.shape[0]):
        ans.update({results[k,0]:results[k,1]})
    dict_results.update({i:ans})
display(pd.DataFrame(dict_results)) 
display(pd.DataFrame(dict({"ROC":claimed})))
print ("Accuray", (claimed.get("true positive") + claimed.get("true negative"))/(claimed.get("true positive") + claimed.get("true negative")+claimed.get("false positive") + claimed.get("false negative" )))
print ("Precision",claimed.get("true positive")/((claimed.get("true positive")+(claimed.get("false positive")))))
print ("Recall",claimed.get("true positive")/((claimed.get("true positive")+(claimed.get("false negative")))))
```

# Results of Identification

```python
def identification(probe):
    similarity_list=list()
    dummy = np.array(data,dtype=float)
    for k in dummy[:,-1]:
        x= getX(float(k)) 
        sub = subtract(x,probe)
        similarity = 0
        for i in range(Ns):
            alpha = alphas[0][i]
            similarity = similarity + (alpha * computeKernal(support_vectors[i],sub))
        similarity_list.append([similarity[0],k])
    
    return similarity_list
```

```python
ident_claimed ={"false positive":0,
         "false negative": 0,
         "true positive":0,
         "true negative":0}
def identificationScore(actual,sim):
    check = sim[0]
    if check < threshold:
        if(sim[1] == actual):
            val=ident_claimed.get("true positive")
            ident_claimed.update({"true positive":val+1})
        else:
            val=ident_claimed.get("false positive")
            ident_claimed.update({"false positive":val+1})
    else:
        if(sim[1]== actual):
            val=ident_claimed.get("false negative")
            ident_claimed.update({"false negative":val+1})
        else:
            val=ident_claimed.get("true negative")
            ident_claimed.update({"true negative":val+1})
```

```python
for i in range(12):
    res = identification(testSet[i,:-1])
    res=pd.DataFrame(np.array(res))
    res =res.sort_values(by=[res.columns[0]])
    identificationScore(testSet[i,-1],np.array(res)[0])
print(ident_claimed)
```
